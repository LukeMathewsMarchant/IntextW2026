## Chapter 8: Anomaly and Insight Discovery

```python
# 1) Logical anomalies

# Donation date before supporter creation date
negative_days = df_prepared[df_prepared['days_to_first_donation'] < 0]
print('Donation before creation date:', len(negative_days))
if len(negative_days) > 0:
    display(negative_days[['supporter_id', 'created_at', 'first_donation_date', 'days_to_first_donation']].head())

# Missing key identifiers
missing_name = df_prepared[(df_prepared['display_name'].isna()) | (df_prepared['display_name'] == '')]
print('Missing display_name:', len(missing_name))

# Unexpected status values
valid_status = {'Active', 'Inactive'}
unexpected_status = df_prepared[~df_prepared['status'].isin(valid_status)]
print('Unexpected status rows:', len(unexpected_status))
if len(unexpected_status) > 0:
    display(unexpected_status[['supporter_id', 'status']].head())
```

```python
# 2) Statistical anomalies for donation timing (IQR method)

donation_days = df_prepared['days_to_first_donation'].dropna()

if len(donation_days) > 0:
    q1 = donation_days.quantile(0.25)
    q3 = donation_days.quantile(0.75)
    iqr = q3 - q1
    lower = q1 - 1.5 * iqr
    upper = q3 + 1.5 * iqr

    outliers = df_prepared[(df_prepared['days_to_first_donation'] < lower) | (df_prepared['days_to_first_donation'] > upper)]

    print(f'IQR lower bound: {lower:.2f}')
    print(f'IQR upper bound: {upper:.2f}')
    print('Outlier rows:', len(outliers))

    if len(outliers) > 0:
        display(outliers[['supporter_id', 'supporter_type', 'acquisition_channel', 'days_to_first_donation']].sort_values('days_to_first_donation').head(15))
else:
    print('No donation timing values available for anomaly check.')
```

```python
# 3) Insight table: where donations appear strongest

insight_table = (
    df_prepared.assign(has_donated=df_prepared['first_donation_date'].notna())
    .groupby(['supporter_type', 'acquisition_channel'])['has_donated']
    .mean()
    .reset_index(name='donation_rate')
    .sort_values('donation_rate', ascending=False)
)

insight_table.head(15)
```

## Chapter 9: Modeling Setup

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score

# Ensure modeling fields exist
model_df = df_prepared.copy()
model_df['has_donated'] = model_df['first_donation_date'].notna().astype(int)
model_df['days_to_first_donation'] = (model_df['first_donation_date'] - model_df['created_at']).dt.days

# A simple tenure feature
max_created = model_df['created_at'].max()
model_df['tenure_days'] = (max_created - model_df['created_at']).dt.days

# Keep rows with valid regression target
reg_df = model_df.dropna(subset=['days_to_first_donation']).copy()

print('Classification rows:', len(model_df))
print('Regression rows:', len(reg_df))
```

## Chapter 10: Multiple Linear Regression (Predict days to first donation)

```python
from sklearn.linear_model import LinearRegression

reg_features = ['supporter_type', 'relationship_type', 'region', 'country', 'status', 'acquisition_channel', 'tenure_days']
X_reg = pd.get_dummies(reg_df[reg_features], drop_first=True)
y_reg = reg_df['days_to_first_donation']

Xr_train, Xr_test, yr_train, yr_test = train_test_split(X_reg, y_reg, test_size=0.25, random_state=42)

lin_reg = LinearRegression()
lin_reg.fit(Xr_train, yr_train)

yr_pred = lin_reg.predict(Xr_test)

print('Linear Regression Metrics')
print('MAE:', round(mean_absolute_error(yr_test, yr_pred), 3))
print('RMSE:', round(np.sqrt(mean_squared_error(yr_test, yr_pred)), 3))
print('R2:', round(r2_score(yr_test, yr_pred), 3))
```

## Chapter 11: Interpretable Effects from Linear Model

```python
coef_table = pd.DataFrame({
    'feature': X_reg.columns,
    'coefficient': lin_reg.coef_
}).sort_values('coefficient', key=np.abs, ascending=False)

coef_table.head(15)
```

## Chapter 12: Decision Tree (Regression)

```python
from sklearn.tree import DecisionTreeRegressor

tree_reg = DecisionTreeRegressor(max_depth=4, random_state=42)
tree_reg.fit(Xr_train, yr_train)

yr_tree_pred = tree_reg.predict(Xr_test)

print('Decision Tree Regression Metrics')
print('MAE:', round(mean_absolute_error(yr_test, yr_tree_pred), 3))
print('RMSE:', round(np.sqrt(mean_squared_error(yr_test, yr_tree_pred)), 3))
print('R2:', round(r2_score(yr_test, yr_tree_pred), 3))
```

## Chapter 13: Classification (Will a supporter donate?)

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier

clf_features = ['supporter_type', 'relationship_type', 'region', 'country', 'status', 'acquisition_channel', 'tenure_days']
X_clf = pd.get_dummies(model_df[clf_features], drop_first=True)
y_clf = model_df['has_donated']

# Use stratify only when each class has at least 2 rows.
class_counts = y_clf.value_counts()
use_stratify = class_counts.min() >= 2 if len(class_counts) > 1 else False
stratify_arg = y_clf if use_stratify else None

Xc_train, Xc_test, yc_train, yc_test = train_test_split(
    X_clf, y_clf, test_size=0.25, random_state=42, stratify=stratify_arg
)

log_clf = LogisticRegression(max_iter=2000)
log_clf.fit(Xc_train, yc_train)
yc_pred_log = log_clf.predict(Xc_test)
yc_prob_log = log_clf.predict_proba(Xc_test)[:, 1]

tree_clf = DecisionTreeClassifier(max_depth=4, random_state=42)
tree_clf.fit(Xc_train, yc_train)
yc_pred_tree = tree_clf.predict(Xc_test)
yc_prob_tree = tree_clf.predict_proba(Xc_test)[:, 1]

def safe_roc_auc(y_true, y_prob):
    return roc_auc_score(y_true, y_prob) if len(pd.Series(y_true).unique()) > 1 else np.nan

def clf_metrics(name, y_true, y_pred, y_prob):
    print(name)
    print('  Accuracy :', round(accuracy_score(y_true, y_pred), 3))
    print('  Precision:', round(precision_score(y_true, y_pred, zero_division=0), 3))
    print('  Recall   :', round(recall_score(y_true, y_pred, zero_division=0), 3))
    print('  F1       :', round(f1_score(y_true, y_pred, zero_division=0), 3))
    auc = safe_roc_auc(y_true, y_prob)
    print('  ROC AUC  :', 'N/A (single class in test set)' if pd.isna(auc) else round(auc, 3))

clf_metrics('Logistic Regression', yc_test, yc_pred_log, yc_prob_log)
clf_metrics('Decision Tree Classifier', yc_test, yc_pred_tree, yc_prob_tree)
```

## Chapter 14: Ensemble Models (Bagging, Boosting, Stacking)

```python
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, StackingClassifier

train_class_counts = pd.Series(yc_train).value_counts()
train_unique_classes = len(train_class_counts)

results = []

if train_unique_classes < 2:
    print('Not enough class variety in training data for ensemble classifiers (need at least 2 classes).')
    print('Class counts in yc_train:')
    print(train_class_counts)
else:
    # Pick a safe CV fold count for stacking based on smallest class size.
    min_class_count = int(train_class_counts.min())
    safe_cv = max(2, min(5, min_class_count))

    rf_clf = RandomForestClassifier(n_estimators=200, random_state=42)
    gb_clf = GradientBoostingClassifier(random_state=42)
    stack_clf = StackingClassifier(
        estimators=[('rf', rf_clf), ('gb', gb_clf)],
        final_estimator=LogisticRegression(max_iter=2000),
        cv=safe_cv
    )

    models = {
        'Random Forest (Bagging)': rf_clf,
        'Gradient Boosting': gb_clf,
        'Stacking': stack_clf
    }

    for name, model in models.items():
        try:
            model.fit(Xc_train, yc_train)
            pred = model.predict(Xc_test)
            prob = model.predict_proba(Xc_test)[:, 1]
            auc = roc_auc_score(yc_test, prob) if len(pd.Series(yc_test).unique()) > 1 else np.nan
            results.append({
                'model': name,
                'accuracy': accuracy_score(yc_test, pred),
                'precision': precision_score(yc_test, pred, zero_division=0),
                'recall': recall_score(yc_test, pred, zero_division=0),
                'f1': f1_score(yc_test, pred, zero_division=0),
                'roc_auc': auc
            })
        except ValueError as e:
            print(f'Skipping {name}: {e}')

pd.DataFrame(results).sort_values('roc_auc', ascending=False, na_position='last') if results else pd.DataFrame()
```

```python
# Optional: quick side-by-side summary for regression models
reg_summary = pd.DataFrame([
    {
        'model': 'Linear Regression',
        'MAE': mean_absolute_error(yr_test, yr_pred),
        'RMSE': np.sqrt(mean_squared_error(yr_test, yr_pred)),
        'R2': r2_score(yr_test, yr_pred)
    },
    {
        'model': 'Decision Tree Regressor',
        'MAE': mean_absolute_error(yr_test, yr_tree_pred),
        'RMSE': np.sqrt(mean_squared_error(yr_test, yr_tree_pred)),
        'R2': r2_score(yr_test, yr_tree_pred)
    }
])

reg_summary
```

## Chapter 15: Evaluation and Model Selection

```python
from sklearn.model_selection import GridSearchCV, StratifiedKFold, cross_val_score

# Validation strategy with safety for tiny classes
class_counts = y_clf.value_counts()
min_class_count = int(class_counts.min()) if len(class_counts) > 0 else 0
safe_splits = max(2, min(3, min_class_count)) if min_class_count >= 2 else None

print('Class counts in y_clf:')
print(class_counts)

# Baseline models for evaluation
eval_models = {
    'Logistic Regression': LogisticRegression(max_iter=2000),
    'Decision Tree Classifier': DecisionTreeClassifier(max_depth=4, random_state=42),
    'Random Forest': RandomForestClassifier(n_estimators=200, random_state=42)
}

cv_results = []
if safe_splits is None:
    print('CV ROC AUC skipped: least populated class has fewer than 2 samples.')
    for name in eval_models.keys():
        cv_results.append({'model': name, 'cv_roc_auc_mean': np.nan, 'cv_roc_auc_std': np.nan})
else:
    cv = StratifiedKFold(n_splits=safe_splits, shuffle=True, random_state=42)
    for name, model in eval_models.items():
        try:
            scores = cross_val_score(model, X_clf, y_clf, cv=cv, scoring='roc_auc', error_score=np.nan)
            cv_results.append({
                'model': name,
                'cv_roc_auc_mean': np.nanmean(scores),
                'cv_roc_auc_std': np.nanstd(scores)
            })
        except ValueError as e:
            cv_results.append({'model': name, 'cv_roc_auc_mean': np.nan, 'cv_roc_auc_std': np.nan})
            print(f'{name} CV skipped: {e}')

cv_table = pd.DataFrame(cv_results).sort_values('cv_roc_auc_mean', ascending=False, na_position='last')
cv_table
```

```python
# Hyperparameter tuning (Random Forest) with safety checks
rf_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [3, 5, None],
    'min_samples_split': [2, 5, 10]
}

train_class_counts = pd.Series(yc_train).value_counts()
train_min_class = int(train_class_counts.min()) if len(train_class_counts) > 0 else 0

if train_min_class >= 2:
    tune_cv = max(2, min(3, train_min_class))
    rf_search = GridSearchCV(
        estimator=RandomForestClassifier(random_state=42),
        param_grid=rf_grid,
        scoring='roc_auc',
        cv=tune_cv,
        n_jobs=-1,
        error_score=np.nan
    )

    rf_search.fit(Xc_train, yc_train)
    best_rf = rf_search.best_estimator_

    print('Best RF Params:', rf_search.best_params_)
    print('Best RF CV ROC AUC:', round(rf_search.best_score_, 3) if pd.notna(rf_search.best_score_) else 'N/A')
else:
    print('Grid search skipped: training data has fewer than 2 samples in a class.')
    best_rf = RandomForestClassifier(n_estimators=200, random_state=42)
    best_rf.fit(Xc_train, yc_train)

best_rf_pred = best_rf.predict(Xc_test)
proba = best_rf.predict_proba(Xc_test)
best_rf_prob = proba[:, 1] if proba.shape[1] > 1 else np.zeros(len(Xc_test))

print('Test Accuracy:', round(accuracy_score(yc_test, best_rf_pred), 3))
print('Test F1:', round(f1_score(yc_test, best_rf_pred, zero_division=0), 3))
print('Test ROC AUC:', round(roc_auc_score(yc_test, best_rf_prob), 3) if len(pd.Series(yc_test).unique()) > 1 else 'N/A')
```

```python
# Fairness check by key groups (region and relationship_type)

eval_df = model_df.loc[Xc_test.index, ['supporter_id', 'region', 'relationship_type']].copy()
eval_df['actual'] = yc_test.values
eval_df['pred'] = best_rf_pred
eval_df['prob'] = best_rf_prob

# Positive prediction rate and recall by group
fairness_region = eval_df.groupby('region', group_keys=False).apply(
    lambda g: pd.Series({
        'count': len(g),
        'positive_prediction_rate': g['pred'].mean(),
        'recall': recall_score(g['actual'], g['pred'], zero_division=0)
    })
).reset_index()

fairness_relationship = eval_df.groupby('relationship_type', group_keys=False).apply(
    lambda g: pd.Series({
        'count': len(g),
        'positive_prediction_rate': g['pred'].mean(),
        'recall': recall_score(g['actual'], g['pred'], zero_division=0)
    })
).reset_index()

print('Fairness by Region')
display(fairness_region.sort_values('positive_prediction_rate', ascending=False))

print('Fairness by Relationship Type')
display(fairness_relationship.sort_values('positive_prediction_rate', ascending=False))

# Simple disparity summary
region_gap = fairness_region['positive_prediction_rate'].max() - fairness_region['positive_prediction_rate'].min() if len(fairness_region) > 1 else 0
relationship_gap = fairness_relationship['positive_prediction_rate'].max() - fairness_relationship['positive_prediction_rate'].min() if len(fairness_relationship) > 1 else 0
print('Positive prediction rate gap (region):', round(region_gap, 3))
print('Positive prediction rate gap (relationship_type):', round(relationship_gap, 3))
```

```python
# Business interpretation helper

base_rate = y_clf.mean()
test_base_rate = yc_test.mean()
pred_rate = pd.Series(best_rf_pred).mean()

print('Business Interpretation')
print(f'- Overall donation base rate in full data: {base_rate:.2%}')
print(f'- Donation base rate in test sample: {test_base_rate:.2%}')
print(f'- Model predicted donation rate in test sample: {pred_rate:.2%}')
print('- Practical use: prioritize supporters with highest predicted probabilities for outreach campaigns.')
print('- If recall is high, fewer likely donors are missed; if precision is high, outreach effort is more efficient.')
print('- Use fairness gap results to check if some regions or relationship groups are being under-selected.')
```

## Chapter 17: Deployment

```python
# 1) Train a production-ready pipeline and save artifact
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
import joblib

prod_features = ['supporter_type', 'relationship_type', 'region', 'country', 'status', 'acquisition_channel', 'tenure_days']
prod_target = 'has_donated'

X_prod = model_df[prod_features].copy()
y_prod = model_df[prod_target].copy()

categorical_features = ['supporter_type', 'relationship_type', 'region', 'country', 'status', 'acquisition_channel']
numeric_features = ['tenure_days']

preprocessor = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_features),
        ('num', 'passthrough', numeric_features),
    ]
)

production_model = Pipeline(
    steps=[
        ('preprocess', preprocessor),
        ('classifier', LogisticRegression(max_iter=2000))
    ]
)

production_model.fit(X_prod, y_prod)

bundle = {
    'model_pipeline': production_model,
    'features': prod_features,
    'target': prod_target,
    'model_name': 'supporters_logistic_pipeline_v1'
}

joblib.dump(bundle, '../artifacts/supporters_production_model.joblib')
print('Saved: ../artifacts/supporters_production_model.joblib')
```

```python
# 2) Create API file for web app deployment (FastAPI)
api_code = """
from fastapi import FastAPI
from pydantic import BaseModel
import pandas as pd
import joblib

app = FastAPI(title='Supporters Donation Propensity API')

bundle = joblib.load('../artifacts/supporters_production_model.joblib')
model = bundle['model_pipeline']
features = bundle['features']
target = bundle['target']

class PredictInput(BaseModel):
    data: dict

@app.get('/health')
def health():
    return {'status': 'ok'}

@app.post('/predict')
def predict(payload: PredictInput):
    row = {f: payload.data.get(f, None) for f in features}
    X = pd.DataFrame([row])
    prob = float(model.predict_proba(X)[0][1])
    pred = int(model.predict(X)[0])
    return {
        'target': target,
        'prediction_label': pred,
        'prediction_probability': round(prob, 4),
        'features_used': features
    }
""".strip()

with open('../artifacts/supporters_deploy_api_example.py', 'w', encoding='utf-8') as f:
    f.write(api_code)

print('Saved: ../artifacts/supporters_deploy_api_example.py')
```
